```python
# Merge with category names
df_merged = df.merge(df_cat, left_on='category_id', right_on='id', how='left')
df_merged = df_merged.drop(columns=['id'])

print(f"Merged shape: {df_merged.shape}")
print(df_merged[['title', 'category_name', 'stars', 'price']].head(3))
```

```python
# Filter out bad rows
df_clean = df_merged[
    (df_merged['title'].notna()) &        # has title
    (df_merged['price'] > 0) &            # has valid price
    (df_merged['stars'] > 0)              # has rating
].copy()

print(f"After filtering: {df_clean.shape}")
print(f"Categories remaining: {df_clean['category_name'].nunique()}")

# Stratified sample: 8000 rows balanced across categories
SAMPLE_SIZE = 8000

df_sampled = (
    df_clean.groupby('category_name', group_keys=False)
    .apply(lambda x: x.sample(
        min(len(x), max(1, int(SAMPLE_SIZE * len(x) / len(df_clean)))),
        random_state=42
    ))
)

# If slightly under 8000, top it up
if len(df_sampled) < SAMPLE_SIZE:
    extra = df_clean[~df_clean.index.isin(df_sampled.index)].sample(
        SAMPLE_SIZE - len(df_sampled), random_state=42
    )
    df_sampled = pd.concat([df_sampled, extra])

df_sampled = df_sampled.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"\nFinal sample shape: {df_sampled.shape}")
print(f"\nCategory distribution (top 10):")
print(df_sampled['category_name'].value_counts().head(10))
```

```python
# Save sampled dataset
df_sampled.to_csv('D:/Semantic-Ecommerce-Recommender/data/raw/amazon_sample.csv', index=False)
print("Sample saved successfully!")
print(f"\nFinal columns: {df_sampled.columns.tolist()}")
```
